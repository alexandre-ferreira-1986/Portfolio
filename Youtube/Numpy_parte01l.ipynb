# Aula 02 - Arrays

**O que são arrays?**

Um ndarray, ou array N-dimensional, é a estrutura de dados principal da biblioteca NumPy em Python.

Representa um array multidimensional e homogêneo de itens de tamanho fixo. No NumPy, as dimensões são referidas como eixos.

Esta estrutura é projetada para operações matemáticas e lógicas eficientes e pode lidar com grandes conjuntos de dados de maneira mais eficaz do que as listas tradicionais do Python.

Os ndarrays suportam uma ampla gama de operações matemáticas e são cruciais para computações numéricas em computação científica, análise de dados e aprendizado de máquina.

---
Os ndarrays no NumPy diferem das listas do Python em termos de homogeneidade e eficiência:

1. **Homogeneidade**: Os ndarrays são homogêneos, o que significa que todos os elementos devem ser do mesmo tipo de dados. Isso contrasta com as listas do Python, que podem conter elementos de tipos diferentes.

2. **Eficiência**: Os ndarrays são mais eficientes em termos de memória e oferecem processamento mais rápido em comparação com as listas do Python. Isso ocorre porque eles são implementados em C e Fortran, permitindo operações otimizadas, especialmente em grandes conjuntos de dados.

Essas diferenças tornam os ndarrays mais adequados para cálculos numéricos onde o desempenho e a utilização de recursos são críticos.

Como copiar um pedaço de array sem alterar o original posteriormente?

---
