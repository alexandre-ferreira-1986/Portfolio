### Checking the quantity of medals

```python
# SQL Query to find countries with most total medals
query_total_medals = """
    SELECT 
        NOC, 
        COUNT(*) AS Total_Medals
    FROM df_participants
    WHERE Medal IS NOT NULL
    GROUP BY NOC
    ORDER BY Total_Medals DESC
    LIMIT 5
"""

# SQL Query to find countries with most gold medals
query_gold_medals = """
    SELECT 
        NOC, 
        COUNT(*) AS Gold_Medals
    FROM df_participants
    WHERE Medal = 'Gold'
    GROUP BY NOC
    ORDER BY Gold_Medals DESC
    LIMIT 5
"""
```

```python
total_medals = pysqldf(query_total_medals)
gold_medals = pysqldf(query_gold_medals)
```

```python
total_medals.head()
```

```python
gold_medals.head()
```

```python
# Plotting
plt.figure(figsize=(12, 6))

# Subplot for countries with most total medals
plt.subplot(1, 2, 1)
plt.bar(total_medals['NOC'], total_medals['Total_Medals'], color='skyblue')
plt.title('Countries with Most Total Medals')
plt.xlabel('Country')
plt.ylabel('Total Medals')

# Subplot for countries with most gold medals
plt.subplot(1, 2, 2)
plt.bar(gold_medals['NOC'], gold_medals['Gold_Medals'], color='gold')
plt.title('Countries with Most Gold Medals')
plt.xlabel('Country')
plt.ylabel('Gold Medals')

plt.tight_layout()
plt.show()
```

### Checking the proportion of men and women by Year

```python
# SQL Query to calculate the distribution of men and women by year
query_gender_distribution = """
    SELECT 
        e.Year, 
        SUM(CASE WHEN p.Sex = 'M' THEN 1 ELSE 0 END) AS Num_Men,
        SUM(CASE WHEN p.Sex = 'F' THEN 1 ELSE 0 END) AS Num_Women
    FROM df_participants AS p
    JOIN df_events AS e 
        ON p.event_id = e.event_id
    GROUP BY e.Year
"""
```

```python
gender_distribution = pysqldf(query_gender_distribution)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.bar(gender_distribution['Year'], gender_distribution['Num_Men'], color='skyblue', label='Men')
plt.bar(gender_distribution['Year'], gender_distribution['Num_Women'], color='pink', bottom=gender_distribution['Num_Men'], label='Women')
plt.title('Distribution of Men and Women by Year')
plt.xlabel('Year')
plt.ylabel('Number of Participants')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(gender_distribution['Year'], gender_distribution['Num_Men'], color='blue', label='Men')
plt.plot(gender_distribution['Year'], gender_distribution['Num_Women'], color='pink', label='Women')
plt.title('Distribution of Men and Women by Year')
plt.xlabel('Year')
plt.ylabel('Number of Participants')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

---

The quantity of women increased a lot. It's almost equal to the men.

Correlation between Participation and Success: You could also analyze the correlation between athlete participation (frequency of participation in events) and their success (number of medals won or type of medal). This analysis could provide insights into the relationship between an athlete's level of participation and their likelihood of achieving success in sports events. For example, you might examine whether athletes who participate in a higher number of events tend to win more medals or if there is a correlation between the consistency of participation and the type of medal won (gold, silver, bronze).

---

```python
query = """
    SELECT 
        Name,
        COUNT(DISTINCT event_id) AS Participation_Count,
        COUNT(Medal) AS Medal_Count,
        SUM(CASE WHEN Medal = 'Gold' THEN 1 ELSE 0 END) AS Gold_Count,
        SUM(CASE WHEN Medal = 'Silver' THEN 1 ELSE 0 END) AS Silver_Count,
        SUM(CASE WHEN Medal = 'Bronze' THEN 1 ELSE 0 END) AS Bronze_Count
    FROM df_participants
    WHERE Medal IS NOT NULL
    GROUP BY Name
"""
```

```python
correlation_df = pysqldf(query)
```

```python
# Calculating correlation between participation count and medal count
correlation = correlation_df['Participation_Count'].corr(correlation_df['Medal_Count'])
```

```python
# Calculating correlation between participation count and gold medal count
gold_correlation = correlation_df['Participation_Count'].corr(correlation_df['Gold_Count'])
```

```python
# Calculating correlation between participation count and silver medal count
silver_correlation = correlation_df['Participation_Count'].corr(correlation_df['Silver_Count'])
```

```python
# Calculating correlation between participation count and bronze medal count
bronze_correlation = correlation_df['Participation_Count'].corr(correlation_df['Bronze_Count'])
```

```python
print("Correlation between Participation Count and Medal Count:", correlation)
print("Correlation between Participation Count and Gold Medal Count:", gold_correlation)
print("Correlation between Participation Count and Silver Medal Count:", silver_correlation)
print("Correlation between Participation Count and Bronze Medal Count:", bronze_correlation)
```

Correlation between Participation and Success: Exploring the relationship between athlete participation and success metrics can uncover insights into performance drivers in sports events. This could involve investigating:
Patterns in athlete participation levels and their impact on performance outcomes, such as medal counts or podium finishes.
Factors influencing athlete participation decisions, including event scheduling, training regimens, or competitive motivations.
Opportunities to optimize athlete performance through targeted training interventions, event selection strategies, or competition planning.

---

```python
import pandas as pd

# Merge df_participants and df_events on event_id
merged_df = pd.merge(df_participants, df_events, on='event_id')

# Group by participant and count the number of medals won
participant_success = merged_df.groupby('Name').agg({'Medal': 'count'}).reset_index()
participant_success.rename(columns={'Medal': 'Success'}, inplace=True)

# Group by participant and count the number of events participated
participant_participation = merged_df.groupby('Name').agg({'event_id': 'nunique'}).reset_index()
participant_participation.rename(columns={'event_id': 'Participation'}, inplace=True)

# Merge participation and success data
df_participation_success = pd.merge(participant_participation, participant_success, on='Name')
```

```python
import seaborn as sns

# Assuming df_participation_success is the dataframe containing relevant data

# Calculate Pearson correlation coefficient
correlation_coefficient = df_participation_success['Participation'].corr(df_participation_success['Success'])

print("Pearson correlation coefficient between Participation and Success:", correlation_coefficient)

# Create a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Participation', y='Success', data=df_participation_success, color='skyblue')
plt.title('Relationship between Participation and Success')
plt.xlabel('Participation')
plt.ylabel('Success')
plt.grid(True)
plt.tight_layout()
plt.show()
```
